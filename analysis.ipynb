```python
import sqlite3
import pandas as pd
con = sqlite3.connect('data.sqlite')
```

```python
cursorObj = con.cursor()

cursorObj.execute('SELECT * FROM AlexaData')
rows = cursorObj.fetchall()
```

```python
print(len(rows), "total event listeners recorded.")
```

```python
cursorObj.execute('SELECT DISTINCT(url) FROM AlexaData')
rows = cursorObj.fetchall()
for row in rows:
    print(row[0])
```

```python
cursorObj.execute("SELECT * FROM AlexaData WHERE nativeEvent = '0'")
rows = cursorObj.fetchall()
print(len(rows), " non-native events")
```

```python
cursorObj.execute("SELECT DISTINCT(type) FROM AlexaData WHERE nativeEvent = '0'")
rows = [x[0] for x in cursorObj.fetchall()]
for row in rows:
    print(row)
```

```python
cursorObj.execute("SELECT * FROM AlexaData WHERE nativeEvent = '1'")
rows = cursorObj.fetchall()
print(len(rows), "native event listeners")
```

```python
nonNativePercent = (29687/(171384+29687))*100
print(nonNativePercent,"% non native event listeners")
print(100 - nonNativePercent,"% native event listeners")
```

```python
cursorObj.execute("SELECT COUNT(*) FROM AlexaData GROUP BY url")
rows = [x[0] for x in cursorObj.fetchall()]
print((sum(rows) / len(rows)), "average interactive elements per url")
```

```python
cursorObj.execute("SELECT COUNT(*) FROM AlexaData GROUP BY url WHERE nativeEvent = ''")
rows = [x[0] for x in cursorObj.fetchall()]
print((sum(rows) / len(rows)), "average interactive elements per url")
```
